### **Top Tags**

The column contains words that are used to describe the restaurants.

We will be finding that are commonly used for Restaurants having food ratings greater then 4 out of 5

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

Here, we will perform below actions:
1. Filter out the rows having null values in column `top_tag`
2. Filter the restaurants have ratings more then 4 in the column `food`
3. Create a word cloud from the above data to get the most commonly used food tags

```python
top_tags=df[df['top_tags'].notna()]
top_tags=top_tags[top_tags['food']>4]
top_tags=top_tags['top_tags'].reset_index(drop=True
                                          )
```

```python
top_tags.head(1)
```

Here, we can seet that the values in `top_tags` has comma seperated values.

In order to generate a word cloud , we will split these values as single values and save it in a list.

Will be using `.split()` function inside a `for` loop.

```python
list_new=[]
list_new1=[]
for i in range(len(top_tags)):
  if ',' in top_tags[i]:
    list_new=(top_tags[i].split(','))
    for i in range(0,len(list_new)):
      list_new1.append(list_new[i])
  else:
    list_new=top_tags[i]
    list_new1.append(list_new)
    
```

We will create a word cloud using `wordcloud` library.

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

#convert list to string and generate
unique_string=(" ").join(list_new1)
wordcloud = WordCloud(width = 1000, height = 500).generate(unique_string)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
```

As we can see, the commonly used food tags for Resturants having food ratings more then 4 are : **Cheap Eats, Vegetarian Friendly, Mid Range, Italian Pizza**

So if a restuarants owner wishes to improve the restaurant's business, they can include the above food options in their menu.

### **3. Are the popular restaurants affordable?**



We will use the DataFrame created from above question and exclude the line items that do not have details of the price range.

Sort the data on `Count of awards` in descending order.

Replace the **€** from the `price range` to blank and calculate the mean of the price range.
 

```python
awards_data=awards_data[awards_data['price_range'].notna()]
awards_data_price=awards_data.sort_values(by='Count of awards',ascending=False).head(50)
```

```python
awards_data_price=awards_data_price.sort_values(by='price_range')
```

Here, we will replace € symbol and '-' from the `price range` column to blank

```python
awards_data_price['price_range']=awards_data_price['price_range'].str.replace('€','')
awards_data_price['price_range']=awards_data_price['price_range'].str.replace('-',',')
```

```python
awards_data_price=awards_data_price.reset_index()
awards_data_price
```

Here the values in the `price range` column are in string format so will split the values as they are comma seperated and save them in list.

After saving values in list, we will calculate its mean and create a new dataframe of the data

```python
listo=[]
length=[]
restaurant=[]
country=[]
Count_of_awards=[]
excellent=[]
for i in range(len(awards_data_price)):
  listo=awards_data_price['price_range'].iloc[i].split(",")
  length.append(listo)
  restaurant.append(awards_data_price['restaurant_name'].iloc[i])
  country.append(awards_data_price['country'].iloc[i])
  Count_of_awards.append(awards_data_price['Count of awards'].iloc[i])
  excellent.append(awards_data_price['excellent'].iloc[i])
```

```python
price_data={'restaurant_name':restaurant,'price_range':length,'country':country,'Count_of_awards':Count_of_awards,'excellent':excellent}
price_data_df=pd.DataFrame(price_data)
```

```python
price_data_df
```

Calculating the mean of the `price range` values.

```python
import statistics
value_mean=[]
for i in range(0,len(price_data_df)):
  value=[]
  value=price_data_df['price_range'].iloc[i]
  value_mean.append((int(value[0])+int(value[1]))/2)
print(value_mean)
```

```python
price_data_df['price_range_mean']=value_mean
```

```python
price_data_df
```

Now, we will convert the datatype of `Count_of_awards` and `excellent` to int for the purpose of sorting and ploting the values on graph

```python

price_data_df['Count_of_awards'] = pd.to_numeric(price_data_df['Count_of_awards'])
price_data_df['excellent'] = pd.to_numeric(price_data_df['excellent'])
```

```python
price_data_df=price_data_df.sort_values(by='Count_of_awards',ascending=False).head(20)
price_data_df
```

```python
fig=px.scatter(price_data_df,
               x='price_range_mean',
               y='Count_of_awards',
               color='country',
               size='excellent',
               title='Price range of Award Winning Restaurants')
fig.show()
```

From the above scatter plot, we can see that most of the restaurants with highest number of awards have a price range between **5-20 €**, which is quite affordable.
